## Step 3: Calculations

Percentage of budget used:
= (Total_Expenses/Budget_Amount) * 100

Percentage of available money used:
= Total_Expenses/ Available_Money * 100

Total Expenses:
= (Sum of Expenses)

Net Balance
= Available_Money – Total_Expenses
